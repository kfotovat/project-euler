<b>Problem 3: Largest Prime Factor</b>

The prime factors of 13195 are 5, 7, 13 and 29.

What is the largest prime factor of the number 600851475143 ?

```python
def naive_prime_factors(x):
    i = 2
    lead = []
    tail = []
    while i * i <= x:
        if x % i == 0:
            lead.append(i)
            n = x / i
            tail.append(n)
        i += 1
    if tail:
        tail.reverse()
        factors = lead + tail
        return factors
    return lead
```

```python
def so_brute_force(n):
    i = 2
    factors = []
    while i * i <= n:
        if n % i:
            i += 1
        else:
            n //= i
            factors.append(i)
    if n > 1:
        factors.append(n)
    return factors
```

```python
def revised_prime_factors(n):
    i = 2
    factors = []
    while i * i <= n:
        if n % i == 0:
            # do something with factors
            factors.append(i)
            n = n / i
            # keep count of instances for exponential notation
            # instances = 0
            while n % i == 0:
                # instances += 1
                n = n / i
        i += 1
    if n != 1:
        factors.append(n)
    return factors
```

<b>Problem 5: Smallest Multiple</b>

2520 is the smallest number that can be divided by each of the numbers from 1 to 10 without any remainder.

What is the smallest positive number that is evenly divisible by all of the numbers from 1 to 20?

```python
import itertools
```

```python
def small_multiple(i, min, max):
    candidates = []
    m = i + 1
    for x in xrange(min,max,2):
    # for x in itertools.count(100):
        flag = True
        for j in range(2, m):
            if x % j != 0:
                flag = False
                # print x, j
                break
        if flag:
            candidates.append(x)
    return candidates
```

```python
def so_smallest_multiple(n):
    dividends = [x for x in range(2, n+1)] # the numbers by which the program must divide
    result = n # will increase in increments of n
    count = 0 #  result has been found when count == n
    while count < n:
        for x in xrange(len(dividends)):
            if result % dividends[x] == 0:
                count += 1 # increases count for every successful division
            count = 0 # if a division fails, count returns to 0
            result += n # and the result is increased by n
    return dividends
```

<b>Problem 6: Sum square difference</b>

The sum of the squares of the first ten natural numbers is,

1^2 + 2^2 + ... + 10^2 = 385

The square of the sum of the first ten natural numbers is,

(1 + 2 + ... + 10)^2 = 55^2 = 3025

Hence the difference between the sum of the squares of the first ten natural numbers and the square of the sum is 3025 − 385 = 2640.

Find the difference between the sum of the squares of the first one hundred natural numbers and the square of the sum.

```python
sum_squares = sum([x*x for x in range(1, 101)])
long_sum = sum([x for x in range(1, 101)])
squared_sum = long_sum * long_sum
answer = squared_sum - sum_squares
```

<b>Problem 9: Special Pythagorean triplet</b>

A Pythagorean triplet is a set of three natural numbers, a < b < c, for which,

a^2 + b^2 = c^2
For example, 3^2 + 4^2 = 9 + 16 = 25 = 5^2.

There exists exactly one Pythagorean triplet for which a + b + c = 1000.
Find the product abc.

```python
def pythagorean_triplet():
    for a in xrange(1, 1000):
        for b in xrange(1, 1000):
            for c in xrange(1, 1000):
                if a*a + b*b == c*c and ((a + b + c) == 1000):
                    return a*b*c
```

<b>Problem 10: Summation of primes</b>

The sum of the primes below 10 is 2 + 3 + 5 + 7 = 17.

Find the sum of all the primes below two million.

```python
def sum_primes(n):
    primes = []
    for i in xrange(2, n):
        for j in xrange(2, n-1):
            flag = True
            if i%j == 0:
                flag = False
        if flag:
            primes.append(i)
    return sum(primes)        
```

```python
sum_primes(6)
```

```python
def primes_sieve(n):
	size = n/2
	sieve = [1]*size
	limit = int(n**0.5)
	for i in xrange(1, limit):
		if sieve[i]:
			val = 2*i+1
			tmp = ((size-1)-i)/val
			sieve[i+val::val] = [0]*tmp
	return [2] + [2*i+1 for i,v in enumerate(sieve) if v and i>0]
```

```python
import itertools

def is_prime(n):
    factors = xrange(2, int(n**(0.5)+1))
    if any(n % f == 0 for f in factors):
        return False
    else: 
        return True    
    
def list_primes(n):
    return [i for i in itertools.count(2) if is_prime(i)][:n]
```
